```python
import pandas as pd
import matplotlib.pyplot as plt
from pytrends.request import TrendReq

pytrends = TrendReq(hl='es-ES', tz=360)
```

```python
search_terms = ["Sea Level", "Wheater",'Temperatures','Carbon Dioxide','Global Warming']
countries = ["AR", "BR", "PY", "CO", "US"]
```

```python
# Crear un diccionario para almacenar los DataFrames de cada país
trends_data_dict = {}

for country in countries:
    # Construir el payload para cada país en el bucle
    pytrends.build_payload(search_terms, timeframe='today 5-y', geo=country)
    trends_data = pytrends.interest_over_time()
    
    # Almacenar el DataFrame en el diccionario con el código de país como clave
    trends_data_dict[country] = trends_data


# Acceder a los DataFrames por país utilizando los códigos de país como claves
df_argentina = trends_data_dict["AR"]
df_brasil = trends_data_dict["BR"]
df_paraguay = trends_data_dict["PY"]
df_colombia = trends_data_dict["CO"]
df_estados_unidos = trends_data_dict["US"]
```
